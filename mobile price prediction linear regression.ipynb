```python
data1=data.drop(["cpu core","cpu freq","ppi"],axis=1)
```

```python
data1.columns
```

```python
y=data1['Price']  #dependent or target variable
X=data1.drop('Price',axis=1)
```

```python
X
```

```python
y
```

# splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

# Applying model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

#test_size=0.30=r2_score=0.9173404473914227
#test_size=0.35=r2_score=0.9193800921588381
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_test)
```

```python
results=pd.DataFrame(columns=['actual','predicted'])
```

```python
results['actual']=y_test
```

```python
results['predicted']=y_pred
```

```python
results['ID']=results.index
results.head(15)
```

## Elastic Regression

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import ElasticNet

elastic = ElasticNet()

parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}

elastic_regressor = GridSearchCV(elastic, parameters)

elastic_regressor.fit(X_train, y_train)
```

```python
elastic_regressor.best_params_
```

```python
elastic=ElasticNet(alpha=.01)
elastic.fit(X_train,y_train)
y_pred_elastic=elastic.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred_elastic)
```

```python
elastic_Error=mean_squared_error(y_pred_elastic,y_test)
elastic_Error
```

```python
Results= pd.DataFrame(columns=['Actual','Predicted'])
Results['Actual']=y_test
Results['Predicted']=y_pred_elastic
Results=Results.reset_index()
Results['Id']=Results.index
Results.head(10)
```

```python
sns.lineplot(x='Id',y='Actual',data=Results.head(50))
sns.lineplot(x='Id',y='Predicted',data=Results.head(50))
plt.plot()
```
